```python
import pandas as pd
dis = 'Jalna'
prep = dis+'_pre.xls'
temp = dis+'_temp.xls'
df = pd.read_table(prep)
df = df.tail(6)
df.tail(6)
```

```python
dft = pd.read_table(temp)
dft = dft.tail(6)
dft.shape
dft.tail(6)
```

```python
spreadsheet = pd.ExcelFile('District_wise_yield(2002).xlsx')
df2 = spreadsheet.parse('Sheet_name_1')
df2 = df2.iloc[1724:1879,:]
df2.tail(10)
```

```python
i=0
j=0
tvar=0
newDf = pd.DataFrame()
newDf = newDf.reset_index()
for row in df2.itertuples():
    if i>-1 :
        year1 = row[3]
        year = year1 - 1901
        tyear = year1 - 1990
        ind = row[0]
        season = row[4].strip()
        newDf = newDf.append(df2.iloc[i:i+1,:], ignore_index = True, sort=False) 
        newDf.iloc[j-1:j,8:9] = df.loc[year,season]
        newDf.iloc[j-1:j,9:10] = dft.loc[year,season]
        tvar=tvar+1
        j=j+1
    i=i+1
newDf.head(10)
```

```python
opf = 'outputs_excels\output' + dis + '.xlsx'
newDf.to_excel(opf)
```
